# Data Understanding
Customer churn, or customers leaving a bank, is a big deal for both the bank and the people involved. For banks, losing customers means losing money. When customers leave, the bank doesn't just lose what they have in their accounts; they lose all the future money the customers would bring in through fees and interest. Getting new customers costs a lot, so keeping the ones they have is crucial for a bank's success.

Lots of things can make customers leave a bank. Bad customer service, high fees, or even better deals from other banks can all lead people to switch. Sometimes, it's because of problems like security breaches or mistakes in their accounts. Even changes in their own financial situations can make people decide to leave. When they do, it's not just a matter of changing a bank account; it can be a stressful experience, especially if bills and payments are involved.

So, it's not just a matter of numbers for banks. Customer churn affects real people, causing disruptions and stress. Banks need to understand why people leave and work on making their services better to keep their customers happy and financially secure.
